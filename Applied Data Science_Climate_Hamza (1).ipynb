```python
import pandas as pd
import wbgapi as wb
```

```python
indic_GDP = ['NY.GDP.MKTP.CD','SL.UEM.1524.FE.ZS','SL.UEM.1524.MA.ZS','NE.IMP.GNFS.ZS']
cntry_code = ["AUS","FRA",'BGR','BGD','ARG','CHN','BRA','GBR','IND','PAK']
indic_Clim=['EN.ATM.CO2E.PC','EN.ATM.GHGT.KT.CE','EG.ELC.ACCS.ZS','EG.ELC.NGAS.ZS']
dta_GDP  = wb.data.DataFrame(indic_GDP, cntry_code, mrv=7)
dta_Clim  = wb.data.DataFrame(indic_Clim, cntry_code, mrv=7)
#NY.GDP.MKTP.CD    : GDP current in USD
#SL.UEM.1524.FE.ZS	Unemployment, youth female (% )
#SL.UEM.1524.MA.ZS	Unemployment, youth male (%)
#NE.IMP.GNFS.ZS: Total import
#EN.ATM.CO2E.PC:	CO2 emissions (metric tons per capita)
#EN.ATM.GHGT.KT.CE: Greenshouse emissions
#EG.ELC.ACCS.ZS	 Electricity access (% of population)	
#EG.ELC.NGAS.ZS	Electricity production from natural gas sources (% of total)
```

```python
# GDP Table
dta_GDP.columns = [a.replace('YR','') for a in dta_GDP.columns]      
dta_GDP=dta_GDP.stack().unstack(level=1)                             
dta_GDP.index.names = ['Country_Code', 'Year']                           
dta_GDP.columns                                                     
dta_GDP.fillna(0)
dta_GDP.head(15)
```

```python
#Climate Table

dta_Clim.columns = [a.replace('YR','') for a in dta_Clim.columns]      
dta_Clim=dta_Clim.stack().unstack(level=1)                             
dta_Clim.index.names = ['Country_Code', 'Year']                           
dta_Clim.columns                                                     
dta_Clim.fillna(0)
dta_Clim.head(15)
```

```python
dta_Clim.head(15)
```

```python
#Dataframe adjustment
d1=dta_GDP.reset_index()
d2=dta_Clim.reset_index()
d3=d1.fillna(0)
d4=d2.fillna(0)
d4.head(10)
```

```python
#Joining the 2 dataset
dta_join = pd.merge(d3, d4)
dta_join.head(10)
```

```python
import matplotlib.pyplot as plt
# Australia statistics
d5=dta_join[(dta_join['Country_Code']=='AUS')]
d5.describe()
```

**The average access to electricity as a total population is 100% in Australia **

```python
# Bar graph - Imports for Australia
d5.plot(x="Year", y="NE.IMP.GNFS.ZS", kind="bar")
```

```python
# France statistics
d6=dta_join[(dta_join['Country_Code']=='FRA')]
d6.describe()
```

**The average net import in France is higher than that of Australia. The average greenshouse gas emission in Australia is higher than that of France**

```python
# Bar graph - GDP for France
d6.plot(x="Year", y="NY.GDP.MKTP.CD", kind="bar")
```

```python
#Bulgaria statistics
d7=dta_join[(dta_join['Country_Code']=='BGR')]
d7.describe()
```

**In Bulgaria, the average electricity production from natural gas sources as a % of total is lower than that of Australia and France. The average total import is higher than that of Australia and France*

```python
# Scatter Plot - Unemployment of Youth Female for Bulgaria
d7.plot(x="Year", y="SL.UEM.1524.FE.ZS", kind="scatter")
```

```python
#Bangladesh statistics
d8=dta_join[(dta_join['Country_Code']=='BGD')]
d8.describe()
```

**In Bangladesh, the average electricity access to total population is less than 100%. The average greenhouse gas emission is higher than that of Australia, France and Bulgaria** 

```python
# Line Graph - Unemployment of Youth male for Bangladesh
plt.plot(d8["Year"], d8["SL.UEM.1524.MA.ZS"])
plt.xlabel("Year")
plt.ylabel("SL.UEM.1524.MA.ZS")
plt.show()
```

```python
#Argentina statistics
d9=dta_join[(dta_join['Country_Code']=='ARG')]
d9.describe()
```
